```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data_frame = pd.read_csv("/content/sample_data/401 Sunset Avenue, Windsor, ONT N9B-3P4 - Sheet1.csv")
print(data_frame)

plt.xlabel("area (square feet)")
plt.ylabel("price (US$)")
plt.scatter(data_frame.area,data_frame.price, color = "green", marker = "+")
plt.plot(data_frame.area,data_frame.price, color = "green", marker = "+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(data_frame[['area']],data_frame.price)
reg.predict([[10000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
